```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r'/content/gdrive/My Drive/Cars_Data.xlsx')
```

```python
display(pd.DataFrame(df))
```

```python
df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Car_Model', y='Cost', data=df)
plt.title('Car Models vs. Costs')
plt.xticks(rotation=45)
plt.xlabel('Car Model')
plt.ylabel('Cost (in INR)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Car_Model', hue='Max_Selling_City', data=df)
plt.title('Car Models vs Max Selling City')
plt.xticks(rotation=45)
plt.xlabel('Car Model')
plt.ylabel('Count')
plt.legend(title='Max Selling City', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Car_Model', hue='Gender_of_buyers', data=df)
plt.title('Car Models vs Gender of Buyers')
plt.xticks(rotation=45)
plt.xlabel('Car Model')
plt.ylabel('Count')
plt.legend(title='Gender', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Profession_of_buyers', y='Cost', data=df)
plt.title('Profession of Buyers vs Cost')
plt.xticks(rotation=45)
plt.xlabel('Profession of Buyers')
plt.ylabel('Cost (in INR)')
plt.show()
```
